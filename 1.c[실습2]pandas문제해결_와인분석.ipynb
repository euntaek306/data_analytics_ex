```python
# 별점 시스템을 만들려고 한다. 따라서 별점에 대한 데이터가 필요하다. 
# 별점은 1,2,3 즉 3개로 만들것이다. 포인트가 95점 이상이면 3점, 85점 이상이면 2점, 나머지는 1점으로 할 것이다. 
# 리뷰데이터를 통해 각 데이터의 별점을 구하시오.
```

```python
df.columns
```

```python
#*    1점
#**   2점
#***  3점
#포인트가 95점 이상이면 3점, 85점 이상이면 2점, 나머지는 1점으로 할 것이다.
# points >= 95 -> 3점
# points >= 85 -> 2점
# 나머지는 1점
```

```python
#df['points']으로 별점 판단하기
#별점 판단하는 함수 정의하고, apply 함수 적용
#1. 함수 정의,  테스트
#2. apply()에 적용하고 df에 적용

#정의부
#def 이름(변수):
    #조건:
        #return 통해 메모리에 기억하고 아래 호출부로 출력, return을 만나는 순간 그 데이터를 호출부에 데이터와 제어를 넘김
def check_star(point):
    # points >= 95 -> 3점
    if point >=95:
        return '***'
    elif point >=85:
        return '**'
    else:
        return '*'
        
    # points >= 85 -> 2점
    # 나머지는 1점
```

```python
# 호출부
# 이름(변수)
# 하나의 값(스칼라 값)으로 테스트
point = 6
check_star(point)
```

- apply()에 적용하고 df에 적용

```python
# 2. apply에 적용 -> 컨테이너 자료형.apply(함수)
df['points'].apply(check_star)
```

```python
# 3. df에 적용, star_point
df['star_point']=df['points'].apply(check_star)
df.head()
```
